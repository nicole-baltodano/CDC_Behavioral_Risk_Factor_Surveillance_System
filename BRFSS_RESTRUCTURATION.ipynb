```python
# Reconnect to the database
conn = sqlite3.connect('brfss_data.db')

# Query the data
query = "SELECT * FROM brfss_data WHERE Year = 2011 AND Locationdesc = 'Alabama';"
filtered_data = pd.read_sql_query(query, conn)

# Close the connection
conn.close()
```

```python
# Find unique values for each column in the dataset
unique_values = {col: filtered_data[col].nunique() for col in filtered_data.columns}

# Display the unique values count for each column
for column, count in unique_values.items():
    print(f"{column}: {count} unique values")
```

```python
# Identify columns with a single unique value, excluding 'Year' and 'Locationdesc'
redundant_columns = [col for col in filtered_data.columns if filtered_data[col].nunique() == 1 and col not in ['Year', 'Locationdesc']]

# Drop the redundant columns
data_cleaned = filtered_data.drop(columns=redundant_columns)

# Display the dropped columns
print(f"Dropped columns: {redundant_columns}")

# Show information about the cleaned DataFrame
print(data_cleaned.info())

```

```python
# Drop columns statistics
columns_to_drop = [ 
    'Display_order', 
    'Data_value',
    'Confidence_limit_Low',
    'Confidence_limit_High'
]

# Drop the columns
data_cleaned = data_cleaned.drop(columns=columns_to_drop)
```

```python
# Display the cleaned DataFrame info
data_cleaned.head()
```

```python
# Select only object (categorical) columns
object_columns = data_cleaned.select_dtypes(include=['object']).columns

# Find and display the actual unique values for each object column
unique_values = {col: data_cleaned[col].unique() for col in object_columns}

# Print the unique values for each object column
for column, values in unique_values.items():
    print(f"Column: {column}")
    print(f"Unique Values: {values}\n")

```

```python
# Filter out rows where Data_Value_Footnote_Symbol has '*', '***'
data_cleaned= data_cleaned[~data_cleaned['Data_Value_Footnote_Symbol'].isin(['*', '***'])]

# Drop the Data_Value_Footnote_Symbol and Data_Value_Footnote columns
data_cleaned = data_cleaned.drop(columns=['Data_Value_Footnote_Symbol', 'Data_Value_Footnote'])

# Display the cleaned DataFrame information
print("Filtered and Cleaned DataFrame:")
print(data_cleaned.info())
```

```python
#Disease =diabetes
diabetes_data = data_cleaned[(data_cleaned['Topic'] == 'Diabetes')]
```

```python
diabetes_data.to_csv('cleaned.csv')
```

```python
diabetes_data.head(10)
```

```python
# Pivot the data to group by Response and make each Break_Out_Category a column
grouped_data = diabetes_data.pivot_table(
    index='Response',
    columns=['Break_Out_Category'],  # Remove Sample_Size from here
    values=['Break_Out', 'Sample_Size'],  # Aggregate these values
    aggfunc={'Break_Out': 'first', 'Sample_Size': 'sum'}
)

# Flatten the MultiIndex columns for easier viewing
grouped_data.columns = [f'{col[0]}_{col[1]}' for col in grouped_data.columns]

# Reset index for better readability
grouped_data = grouped_data.reset_index()
```
